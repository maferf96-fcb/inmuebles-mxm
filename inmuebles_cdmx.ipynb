# **Demandas de DEVS**

## **Crear columnas numéricas**

###valor_mensual

```python
datos = pd.read_csv(url, sep = ";")
datos.shape
```

```python
datos['Valor_mensual'] = datos['Valor'] + datos['Condominio']
datos.head()
```

###valor_anual

```python
datos['Valor_anual'] = datos['Valor_mensual'] * 12 + datos['Impuesto']
datos.head()
```

##**Crear columnas categóricas**

###Descripcion

```python
datos['Descripcion'] = datos['Tipo'] + ' en la colonia ' + datos['Colonia'] + ' con ' + datos['Habitaciones'].astype(str) + ' cuartos(s) y ' + datos['Garages'].astype(str) + ' plazas de estacionamiento'
datos.head()
```

```python
datos.iloc[500].Descripcion
```

###Tiene_suite

```python
datos['Tiene_suite'] = datos['Suites'].apply(lambda x: 'Si' if x > 0 else 'No')
```
